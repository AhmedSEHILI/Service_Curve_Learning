```python
import matplotlib.pyplot as plt
```

```python
def plot_piecewise(ax, P, label=None, lw=2):
    T = [p[0] for p in P]
    Y = [p[1] for p in P]
    ax.plot(T, Y, label=label, linewidth=lw)




def line_rt(T_end, r):
    """Retourne les deux points de la droite y = r*t tronqu√©e sur [0, T_end]."""
    return [(0.0, 0.0), (float(T_end), float(r*T_end))]

```

```python
def cuts_with_line(A, r):
    n = len(A) - 1
    for i in range(1, n):
        T0, Y0 = A[i]
        T1, Y1 = A[i+1]
        v0 = r*T0 - Y0
        v1 = r*T1 - Y1
        if v0 == 0 or v1 == 0 or v0 * v1 < 0:
            return True
            
    return False
```

```python
def minplus_conv_linear(A, r):
    assert len(A) >= 2
    assert r > 0

    T0, A0 = A[0]
    TN, AN = A[-1]
    if T0 != 0.0 or A0 != 0.0:
        raise ValueError("A doit commencer en (0,0)")

    cuts = cuts_with_line(A, r)  
    if not cuts:
        T1, A1 = A[1]
        if r*T1 < A1:
            return [(0.0, 0.0), (float(TN), float(r*TN))]
        else:
            return [(float(T), float(Y)) for (T, Y) in A]

    out = [(0.0, 0.0)]
    M = 0.0

    for i in range(1, len(A)):
        Ti, Ai = A[i]
        M = min(M, Ai - r*Ti)
        C = r*Ti + M
        K = Ti

        if C != Ai:
            Tim1, Aim1 = A[i-1]
            K = ((Ai - Aim1) / r) + Tim1
            C = Ai

        out.append((float(K), float(C)))

    return out
```

```python
A = [
(0.0, 0.0),
(1.0, 2.0), 
(3.0, 2.0), 
(4.0, 4.0), 
(5.0, 4.0), 
(6.0, 8.0), 
(10.0, 8.0), 
]
r = 1.3
    
    
C = minplus_conv_linear(A, r)
L = line_rt(A[-1][0], r)
    
    
fig, ax = plt.subplots(figsize=(8, 5))
plot_piecewise(ax, A, label="A(t) ‚Äî arriv√©es", lw=2)
plot_piecewise(ax, L, label=f"S(t) = {r}¬∑t", lw=2)
plot_piecewise(ax, C, label="A*S", lw=3)
    
    
ax.set_xlabel(" t")
ax.set_ylabel("flux")
ax.set_title("Conv min-plus")
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
def dev_h(A, D):
    Y = sorted(set([Ai for _, Ai in A] + [Dj for _, Dj in D]))

    i = 0
    j = 0
    N = len(A) - 1
    M = len(D) - 1
    Delta = 0.0

    for y in Y:
        while i < N and A[i][1] < y:
            i += 1

        Ai = A[i][1]; Ti = A[i][0]
        if Ai == y:
            if i < N and A[i+1][1] == y:
                tA = A[i+1][0]
            else:
                tA = Ti
        else:
            Tim1, Aim1 = A[i-1]
            tA = Tim1 + (y - Aim1) * (Ti - Tim1) / (Ai - Aim1)

        while j < M and D[j][1] < y:
            j += 1

        if j == M and D[j][1] < y:
            return float('inf')

        Dj = D[j][1]; Kj = D[j][0]
        if Dj == y:
            if j < M and D[j+1][1] == y:
                tD = D[j+1][0]
            else:
                tD = Kj
        else:
            Kj_m1, Dj_m1 = D[j-1]
            tD = Kj_m1 + (y - Dj_m1) * (Kj - Kj_m1) / (Dj - Dj_m1)

        Delta = max(Delta, tD - tA)

    return Delta
```

```python
D = [
    (0.0, 0.0),
    (1.0, 1.0),
    (2.0, 1.0),
    (3.0, 2.0),
    (5.0, 2.0),
    (6.0, 4.0),
    (8.0, 4.0),
    (10.0, 8.0),
    (11.0, 8.0),
]

Delta = dev_h(C, D)
print("DevH(A, C) =", Delta)

fig, ax = plt.subplots(figsize=(8, 5))
plot_piecewise(ax, A, label="A(t)", lw=2)
plot_piecewise(ax, L, label=f"S(t) = {r}¬∑t", lw=2)
plot_piecewise(ax, C, label="A * S", lw=3)
plot_piecewise(ax, D, label="D(t)", lw=2)

ax.set_xlabel("t")
ax.set_ylabel("flux ")
ax.set_title("Conv")
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

```python

# C d√©cal√©e vers la droite de Delta
C_shift = [(t + Delta, y) for (t, y) in C]

fig, ax = plt.subplots(figsize=(8, 5))
plot_piecewise(ax, A, label="A(t)", lw=2)
plot_piecewise(ax, D, label="D(t)", lw=2)
plot_piecewise(ax, C_shift, label=f"C(t) d√©cal√©e", lw=3)

ax.set_xlabel("t")
ax.set_ylabel("flux ")
ax.set_title("C d√©cal√©e")
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
def beta_rt(T_end, r, Delta):
    if Delta >= T_end:
        return [(0.0, 0.0), (float(T_end), 0.0)]
    if Delta <= 0.0:
        return [(0.0, 0.0), (float(T_end), float(r*T_end))]
    return [(0.0, 0.0), (float(Delta), 0.0), (float(T_end), float(r*(T_end - Delta)))]

T_end = max(A[-1][0], D[-1][0])
Beta = beta_rt(T_end, r, Delta)

fig, ax = plt.subplots(figsize=(8, 5))
plot_piecewise(ax, A, label="A(t)", lw=2)
plot_piecewise(ax, D, label="D(t)", lw=2)
plot_piecewise(ax, Beta, label=f"beta", lw=3)

ax.set_xlabel("t")
ax.set_ylabel("flux cumul√©")
ax.set_title("A, D et beta")
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```
